## Final code

```python
import cv2

face_cap = cv2.CascadeClassifier(
    "C:/Users/ashwa/anaconda3/Lib/site-packages/cv2/data/haarcascade_frontalface_default.xml"
)

video_cap = cv2.VideoCapture(0)

while True:
    
    ret, video_data = video_cap.read()
    if not ret:
        print("Camera not detected")
        break
    
    gray = cv2.cvtColor(video_data, cv2.COLOR_BGR2GRAY)

    faces = face_cap.detectMultiScale(
        gray,
        scaleFactor = 1.1,
        minNeighbors = 5,
        minSize = (30, 30)
    )

    for (x, y, w, h) in faces:
        cv2.rectangle(video_data, (x, y), (x+w, y+h), (0,255,0), 2)

    cv2.imshow("Video_live", video_data)

    if cv2.waitKey(10) == ord("a"):
        break

video_cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
from IPython.display import display, Image, clear_output

face_cap = cv2.CascadeClassifier(
    "C:/Users/ashwa/anaconda3/Lib/site-packages/cv2/data/haarcascade_frontalface_default.xml"
)

video_cap = cv2.VideoCapture(0)

while True:
    ret, frame = video_cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cap.detectMultiScale(gray, 1.1, 5)

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0,255,0), 2)

    # Convert frame to JPEG and display in notebook
    _, jpeg = cv2.imencode('.jpg', frame)
    display(Image(data=jpeg.tobytes()))
    clear_output(wait=True)

    if cv2.waitKey(1) & 0xFF == ord('a'):
        break

video_cap.release()
cv2.destroyAllWindows()
```
